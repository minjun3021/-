```python
import csv
import operator
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(16,9))
f=open('학교급 및 시도별 학생 1인당 월평균 사교육비.csv',encoding='cp949')
data= csv.reader(f)
dataD = {}
for line in data:
  
    if line[1] == "일반고[만원]" and line[0] != "세 종":
        if dataD.get(line[0]) == None:
            dataD[line[0]] = 0
        dataD[line[0]] += float(line[4])-float(line[3])
dataSorted={}        
dataSorted=dict(sorted(dataD.items(), key=operator.itemgetter(1)))

plt.title("도별 일반고 학생 1인당 (2009~2018) 평균 사교육비 증가 액수",size=30)
plt.bar(dataSorted.keys(),dataSorted.values(),color='skyblue',width=0.5)
plt.xticks(rotation=45,size=15)
plt.yticks(size=15)


#-------------------------------------------------
plt.rc('font',family='Malgun Gothic')
plt.figure(figsize=(16,9))
f=open('학생성적 순위별 학생 1인당 월평균 사교육비.csv',encoding='cp949')
data= csv.reader(f)
saDict={}
for line in data:
    if(line[0]=="사교육비"):
        print(line[1])
        if(saDict.get(line[1])==None):
            
            saDict[line[1].split('[')[0]]=0
            saDict[line[1].split('[')[0]]+=float(line[3])


plt.title("학생성적 순위별 학생 1인당 월평균 사교육비[만원]",size=30)
plt.bar(saDict.keys(),saDict.values(),color='r',width=0.5)
plt.xticks(rotation=20,size=15)
plt.yticks(size=15)
            
            
```
